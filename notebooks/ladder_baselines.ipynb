```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../models/ladder_train_total_loss.txt', 'r') as filehandle:  
    lines = filehandle.readlines()
    train_total_loss = [float(line[:-1]) for line in lines]
with open('../models/ladder_train_l_loss.txt', 'r') as filehandle:  
    lines = filehandle.readlines()
    train_l_loss = [float(line[:-1]) for line in lines]
with open('../models/ladder_train_u_loss.txt', 'r') as filehandle:  
    lines = filehandle.readlines()
    train_u_loss = [float(line[:-1]) for line in lines]
```

```python
sns.set_style('darkgrid')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 5))
ax_l = sns.lineplot(x=np.arange(len(train_l_loss)), y=train_l_loss, ax=ax[0])
ax_l.set_title('Supervised learning curve')
ax_l.set_xlabel('number of iterations')
ax_l.set_ylabel('labeled loss')
ax_l.plot()
ax_u = sns.lineplot(x=np.arange(len(train_u_loss)), y=train_u_loss, ax=ax[1])
ax_u.set_title('Unsupervised learning curve')
ax_u.set_xlabel('number of iterations')
ax_u.set_ylabel('unlabeled loss')
ax_u.plot()
```
